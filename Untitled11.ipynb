```python
selected_features = [
'Heart Rate', 'Respiratory Rate', 'Body Temperature',
'Oxygen Saturation', 'Systolic Blood Pressure', 'Diastolic Blood Pressure',
'Age', 'Gender', 'Derived_BMI', 'Derived_HRV', 'Risk Category'
]

df = df[selected_features]

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(data=df[['Heart Rate','Respiratory Rate','Body Temperature',
                     'Oxygen Saturation','Systolic Blood Pressure',
                     'Diastolic Blood Pressure','Derived_BMI','Derived_HRV']])

plt.title("Outlier Detection using Box Plot")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

avg_hr = df.groupby('Gender')['Heart Rate'].mean()

avg_hr.plot(kind='bar')

plt.title("Average Heart Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Heart Rate")
plt.show()
```

```python
df['Risk Category'].value_counts().plot(kind='bar')

plt.title("Risk Category Distribution")
plt.xlabel("Risk Category")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix of Vital Signs")
plt.show()
```

```python
sns.pairplot(df[['Heart Rate','Respiratory Rate','Body Temperature',
                 'Oxygen Saturation','Derived_BMI']])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Derived_BMI', y='Systolic Blood Pressure', data=df)

plt.title("BMI vs Systolic Blood Pressure")
plt.xlabel("BMI")
plt.ylabel("Systolic Blood Pressure")
plt.show()
```

```python
df['Risk Category'].value_counts().plot(kind='bar')

plt.title("Risk Category Distribution")
plt.xlabel("Risk Category")
plt.ylabel("Number of Patients")
plt.show()
```

```python
sns.boxplot(x='Risk Category', y='Age', data=df)

plt.title("Age Distribution by Risk Category")
plt.show()
```
