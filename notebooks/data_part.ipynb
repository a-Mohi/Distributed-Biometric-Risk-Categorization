```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("human_vital_signs_dataset_2024.csv")  # update path as needed

# Split dataset into 5 parts
part1, part2, part3, part4, part5 = np.array_split(df, 5)

# Check sizes
print("Part 1 shape:", part1.shape)
print("Part 2 shape:", part2.shape)
print("Part 3 shape:", part3.shape)
print("Part 4 shape:", part4.shape)
print("Part 5 shape:", part5.shape)

# Save the distributed partitions
part1.to_csv("vital_signs_node11.csv", index=False)
part2.to_csv("vital_signs_node22.csv", index=False)
part3.to_csv("vital_signs_node33.csv", index=False)
part4.to_csv("vital_signs_node44.csv", index=False)
part5.to_csv("vital_signs_node55.csv", index=False)
```
