```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# ─────────────────────────────────────────────
# LOAD CLEANED DATA
# ─────────────────────────────────────────────
df = pd.read_csv("data_cleaned1.csv")  # update path as needed

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
pd.DataFrame(X).head()
```

```python
le_gender = LabelEncoder()
X['gender'] = le_gender.fit_transform(X['gender'])
```

```python
pd.DataFrame(X).head()
```

```python
# STEP 3: LABEL ENCODE risk_category (ordinal target)
risk_mapping = {
    "Low Risk"      : 0,
    "High Risk"     : 1
}

y_encoded = y.map(risk_mapping)
y_encoded.name = "risk_category"

print("\n--- After Label Encoding risk_category ---")
print(f"Mapping : {risk_mapping}")
print(f"Unique values: {sorted(y_encoded.unique().tolist())}")
print(y_encoded.value_counts().sort_index())


Mapping : {'Low Risk': 0,  'High Risk': 1}
```
